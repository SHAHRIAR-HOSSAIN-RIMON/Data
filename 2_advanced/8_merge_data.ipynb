```python
df5=pd.read_csv('https://lukeb.co/software_csv',index_col='job_posted_month')
df5
```

```python
df_merged=df5.merge(df5,on='job_posted_month')
```

```python
top_5=(df_merged.sum()
       .sort_values(ascending=False)
       .head(5)
       .index.
       to_list())
top_5
```

```python
df_merged[top_5].plot(kind='line')
plt.title('monthly job postings')
plt.xlabel('Month')
plt.ylabel('Number of postings')
plt.xticks(rotation=45)
plt.legend(loc='upper center')
plt.show()
```
