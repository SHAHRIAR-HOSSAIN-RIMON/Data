```python
%pip install dataset
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

```

```python

dataset =load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df_usa=df[df['job_country']=='United States'].copy()
```

```python
df_usa
```

```python
df_usa['job_posted_month']=df_usa['job_posted_date'].dt.strftime('%B')
df_usa
```

```python
df_pivot=df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
```

```python
df_pivot.reset_index(inplace=True)
df_pivot['month_number']=pd.to_datetime(df_pivot['job_posted_month'],format='%B').dt.month
df_pivot
```

```python
df_pivot
```

```python
df_pivot.sort_values('month_number',inplace=True)
df_pivot.set_index('job_posted_month',inplace=True)
df_pivot.drop(columns='month_number',inplace=True)
```
