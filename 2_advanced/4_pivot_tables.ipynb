```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
dataset =load_dataset('lukebarousse/data_jobs')
df =dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_title_short',aggfunc='size')
```

```python

df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc='mean')
```

```python
df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
```

```python
top_countries=df['job_country'].value_counts().head(10)
top_countries
```

```python
top_countries=df['job_country'].value_counts().head(10)
df_job_country_salary=df.pivot_table(index='job_country',values='salary_year_avg',columns='job_title_short',aggfunc='median')
```

```python
df_job_country_salary
```

```python
df_job_country_salary = df_job_country_salary.loc[top_countries.index]
```

```python
df_job_country_salary
```

```python
job_titles = df_job_country_salary.columns.tolist()
df_job_country_salary = df_job_country_salary[job_titles].head(5)
```
